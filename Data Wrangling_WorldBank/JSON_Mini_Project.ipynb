# 2.Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
#The column 'mjtheme_namecode' is nested. This will need to be flatterned before we can count major project themes.
#A json string is needed to flattern this
#first is to import json_normalize
```

```python
from pandas.io.json import json_normalize
```

```python
#load the json file and save as a json string
json_str = json.load((open('world_bank_projects.json')))
```

```python
#flatterns the column mjtheme_namecode: json_str
json_str = json_normalize(json_str,'mjtheme_namecode')
json_str.head(10)
```

```python
# it appears that some names are missing
# therefore we can count the major themes by utilizing the completely available codes
mjtheme_count = json_str.code.value_counts()

#after counting is done, we call the top 10 major themes and assign to: top10_mjtheme
top10_mjtheme = mjtheme_count.head(10)
top10_mjtheme
```

# 3.In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#handling missing data in this case can be done by filling backward functionality. This will include three steps
# (1) sorting 
# (2) replacing missing data ('') with NaN (will need numpy)
# (3) fill backward

#import numpy as np
import numpy as np
```

```python
# sort by 'code' and 'name'
json_str = json_str.sort_values(['code','name'])
json_str.head(10)                            
```

```python
#the space in the empty cells can be replaced with NaN
json_str.name[json_str['name']==''] = np.nan
json_str.head(10)
```

```python
#fills the missing values in an backward manner
json_str = json_str.fillna(method ='bfill')
json_str.head(10)
```

```python
#now we can count the top 10 major themes by name

top10_mjtheme = json_str.name.value_counts().head(10)
top10_mjtheme
```
